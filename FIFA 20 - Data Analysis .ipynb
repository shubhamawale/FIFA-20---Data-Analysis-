```python
player_salary = df[['short_name','wage_eur']]
```

```python
# heighest payed football player
top_5_payed = player_salary.sort_values('wage_eur', ascending=False).head()
top_5_payed
```

```python
plt.figure(figsize=(8,5))
plt.bar(top_5_payed['short_name'],top_5_payed['wage_eur'], color=['r','g','b','gray','pink'])
plt.show()
```

```python
# dataframe of german player
Germany = df[df['nationality']=='Germany']
Germany.head()
```

```python
# find who is the tallest German player
Germany[Germany['height_cm'] == Germany['height_cm'].max()]
```

```python
# Top 5 tallest German player
Germany.sort_values('height_cm', ascending=False).head()
```

```python
# maximum german player weight
Germany.sort_values(by=['weight_kg'], ascending=False).head()
```

```python
# heighest payed German player
Germany[['short_name','wage_eur']].sort_values('wage_eur',ascending=False).head()
```
